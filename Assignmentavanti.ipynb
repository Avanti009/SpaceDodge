```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
file_path = r'/content/IDZ Digital Data Analyst.xlsx'
df = pd.read_excel(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting age distribution
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.axvline(df['Age'].mean(), color='red', linestyle='dashed', linewidth=2, label='Average Age')
plt.legend()
plt.show()
```


Age Distribution:
The age of individuals ranges from 18 to 60 years.
The average age is around 39 years.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting gender distribution
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
top_occupations = df['Occupation'].value_counts().nlargest(10).index
df_top_occupations = df[df['Occupation'].isin(top_occupations)]

# General Occupation Distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='Occupation', data=df_top_occupations)
plt.title('Top 10 Occupation Distribution')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()

# Average Salary by Top Occupations
avg_salary_by_occupation = df_top_occupations.groupby('Occupation')['Salary'].mean().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_salary_by_occupation.index, y=avg_salary_by_occupation.values)
plt.title('Average Salary by Top 10 Occupations')
plt.xlabel('Occupation')
plt.ylabel('Average Salary')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
top_occupations_salary = df.groupby('Occupation')['Salary'].mean().nlargest(5).index
df_top_occupations_salary = df[df['Occupation'].isin(top_occupations_salary)]

# Average Salary by Top 5 Occupations
plt.figure(figsize=(12, 6))
sns.barplot(x='Occupation', y='Salary', data=df_top_occupations_salary, order=top_occupations_salary)
plt.title('Top 5 Occupations with Highest Average Salary')
plt.xlabel('Occupation')
plt.ylabel('Average Salary')
plt.xticks(rotation=45, ha='right')
plt.show()
```
